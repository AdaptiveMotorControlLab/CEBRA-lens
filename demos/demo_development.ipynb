Step 8: Distance metrics

```python
metric = "cosine"
```

```python
multimodel_interrep = lens.quantification.multibase.MultiModel(
        lens.quantification.distance.Distance(
            data=train_data,
            label=train_label,
            dataset_label=dataset_label,
            metric=metric,
            distance_label = "interrep",
        )
        )
```

```python
interrep_distances_dict = multimodel_interrep.compute(activations_dict)
```

```python
fig = multimodel_interrep.plot(
    interrep_distances_dict,
    title="Inter-representation distance",
)
```

```python
multimodel_intrabin = lens.quantification.mutlibase.MultiModel(
        lens.quantification.distance.Distance(
            data=train_data,
            label=train_label,
            dataset_label=dataset_label,
            metric=metric,
            distance_label = "intrabin",
        )
        )
```

```python
intrabin_distances_dict =  multimodel_intrabin.compute(activations_dict)
```

```python
multimodel_interbin = lens.quantification.multibase.MultiModel(
        lens.quantification.distance.Distance(
            data=train_data,
            label=train_label,
            dataset_label=dataset_label,
            metric=metric,
            distance_label = "interbin",
        )
        )
interbin_distances_dict = multimodel_interbin.compute(activations_dict)
```
